# outliers are present in Standard Error and Trust (Government Corruption)

```python
q1 = df1['Trust (Government Corruption)'].quantile(.25)
q3 = df1['Trust (Government Corruption)'].quantile(.75)

iqr = q3-q1
low = q1-1.5*iqr
up = q3+1.5*iqr

print(low)
print(up)
```

```python
new_df = df1.loc[(df1['Trust (Government Corruption)']<up) & (df1['Trust (Government Corruption)']>low)]
print(len(df1))
print(len(new_df))
```

```python
q1 = new_df['Standard Error'].quantile(.25)
q3 = new_df['Standard Error'].quantile(.75)

iqr = q3-q1
low = q1-1.5*iqr
up = q3+1.5*iqr

print(low)
print(up)
```

```python
new_df2 = new_df.loc[(new_df['Standard Error']<up) & (new_df['Standard Error']>low)]
print(len(new_df))
print(len(new_df2))
```

# Train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = new_df2.drop('Happiness Score', axis=1)
y = new_df2['Happiness Score']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
svm = SVR()
svm.fit(x_train,y_train)
```

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
rfr= RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
gbr= GradientBoostingRegressor()
gbr.fit(x_train,y_train)
```

```python
y_pred1 = svm.predict(x_test)
y_pred2 = lr.predict(x_test)
y_pred3 = rfr.predict(x_test)
y_pred4 = gbr.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
svm1 = metrics.r2_score(y_test,y_pred1)
svm1
```

```python
lr2 = metrics.r2_score(y_test,y_pred2)
lr2
```

```python
rfr3 = metrics.r2_score(y_test,y_pred3)
rfr3
```

```python
gbr4 = metrics.r2_score(y_test,y_pred4)
gbr4
```

# LinearRegression is best
